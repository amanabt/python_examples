```python
# This sample data is taken from one of the Nuclear Magneto Resonance(NMR) experiments
# The 1^{st} and the 2^{nd} axis should have a linear relationship.

sample_data = np.array([[ 12.5 ,   2.85],
                        [ 12.75,   2.92],
                        [ 13.  ,   2.99],
                        [ 13.25,   3.03],
                        [ 13.5 ,   3.08],
                        [ 13.75,   3.14],
                        [ 14.  ,   3.2 ],
                        [ 14.25,   3.25],
                        [ 14.5 ,   3.31],
                        [ 14.75,   3.37],
                        [ 15.  ,   3.43],
                        [ 15.25,   3.49],
                        [ 15.5 ,   3.57],
                        [ 15.75,   3.64],
                        [ 16.  ,   3.67],
                        [ 16.25,   3.72],
                        [ 16.5 ,   3.77],
                        [ 16.75,   3.84],
                        [ 17.  ,   3.89]])
```

```python
linear_fit = np.polyfit(sample_data[:, 0], sample_data[:, 1], deg = 1, full = True)

w = np.linspace(np.ndarray.min(sample_data[:, 0]), np.ndarray.max(sample_data[:, 0]))
B = linear_fit[0][0] * w + linear_fit[0][1]
```

```python
plt.title(r'$B_0$ vs $\omega$')

plt.plot(w, B, '-', label = r'Linear Fit of the Experimental Data')
plt.plot(sample_data[:, 0], sample_data[:, 1], 'o', label = r'Experimental Data')
plt.legend(prop={'size': 14})
plt.show()
```
